# The Receive, Accept, Sample Model and Political Priming

John R. Zaller developed the Receive, Accept, Sample (RAS) model to explain how people respond to political surveys. He published this in ["The Nature and Origins of Mass Opinion"](https://www.cambridge.org/core/books/nature-and-origins-of-mass-opinion/70B1485D3A9CFF55ADCCDD42FC7E926A) in 1992. There are four components to this theory. The first is the reception axiom, which says that the more politically interested people are the more likely they are to receive political messaging. The resistance axiom follows, which says that people resist or reject considerations that do not align with their predispositions or come from sources that they do not trust, like a politician from the opposing party. The third component is the accessibility axiom, which says that more recently thought about considerations take less time to apply and use in a political or voting context. The final component is the response axiom, which says that people sample the relevant considerations based on information that aligns best with them and is most accessible to them. 

The typical voter has medium partisanship and low attention to politics. The RAS model dictates that the probability they will receive information is low, but the probability that they will accept the information they do receive is high. Because they pay less attention to politics, the chance that they will sample this information (look to it as an issue that takes priority in that moment) is higher. This is again, in the context of responding to political surveys. Zaller argued that political surveys are best framed as a measure of considerations that are most important to the person being surveyed in that moment as opposed to a realistic measure of an individual's true preferences, since it does not capture previously held opinions (which Zaller argues, the typical voter does not really have many of). 

The priming hypothesis* refers to the phenomenon that the media's selection, characterization and presentation of political issues prime voters' evaluation of which issues are most salient and indicative of a candidate's main concerns. In other words, the media influences what people think about and what they weigh as most important. As an example, if the media focuses primarily on economic issues, voters are more likely to weigh economic performance in their approval of a political candidate, when asked to do so in a survey. 

*The priming effect was identified [in this 1982 paper](https://www.cambridge.org/core/journals/american-political-science-review/article/abs/experimental-demonstrations-of-the-notsominimal-consequences-of-television-news-programs/6E79DF36E1637BB0E94F8A03DB4EA57B) by Iyengar, Peters, and Kinder, published in the American Political Science Review. # 2020 General Election

First, I'm going to look at the articles published on MSNBC in October 2020, preceding the General Election for that year. Below I define I function to get the "soup" so that I can use it again when getting the MSNBC data for the month preceding midterm election. I elected to use this method over an API because I wanted to try implementing what we learned with web scraping in COMPSS 201. 

```python
#article sitemap stores list of articles for a whole month

#define a function to get the "soup", so I can do this easily again for midterm elections
def get_soup(url):
    response = requests.get(url)
    #line to control for error codes
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        return soup
    else:
        print("URL Not Found")

#tagging these with _ge to denote that these variables refer to info from the month before the general election
msnbc_url_ge = 'https://www.msnbc.com/archive/articles/2020/october'
msnbc_soup_ge = get_soup(msnbc_url_ge)
```

When I opened Inspect Element on the webpage, I saw that the article titles and their respective hyperlinks were tagged as 'a' with an href attribute. I tried copying the CSS selector but it didn't work, so I opted for a more generalized selector below, and just selected all the 'a' tags with 'href' attributes on the page. This worked for my purposes in this notebook because the only things with 'a' tags and 'href' attributes were the article links and titles.

```python
#select only a tags with href attribute (per html)
msnbc_a_ge = msnbc_soup_ge.select('a[href]')
```

Below I initialized a dictionary to store the article titles and their links. I included an if statement in the for loop because I wanted to exclude links that were incomplete or did not link to an article. I made this modification because they initially appeared in the dictionary and were not pertinent to the analysis I planned to do. 

```python
#initialize dictionary to store output
msnbc_articles_ge = {}

#for loop to append the links and titles to the dict
for link in msnbc_a_ge:
        href = link.get('href')
        if "https://" not in href: #getting rid of as many pages as possible that aren't articles (cleaning)
            continue
        title = link.get_text(strip=True) #get the title for the article
        msnbc_articles_ge[href]=title
```

I then converted the dictionary to a pandas DataFrame, since I've only worked with LDA and TF-IDF when the data is stored in a DataFrame. 

```python
#convert dictionary to dataframe to use for LDA and TF-IDF
msnbc_df_ge = pd.DataFrame.from_dict(msnbc_articles_ge, orient='index').reset_index()
msnbc_df_ge = msnbc_df_ge.rename(columns={"index": "url", 0:"title"}) #rename the columns
msnbc_df_ge
```

## LDA

Here I use Latent Dirichlet Allocation to create a list of topics from the corpus of article titles. I'm hoping that the LDA will yield a list of distinct topics that may reveal what issues were determined to be salient for their readers by MSNBC in the month preceding the election. 

When initializing the Count Vectorizer, I set max_df to a high value of .95 to remove terms that appear in the corpus of documents (article headlines) too frequently. I set the min_df to .02 to ignore the terms that have a document frequency less than .02. I had to change the min_df a few times to get an output from my topic model that had enough terms. Documentation for the function that I referred to can be found [here](https://scikit-learn.org/stable/modules/generated/sklearn.feature_extraction.text.CountVectorizer.html). I also referred to code from a lab I did a few years ago in undergrad. 

```python
#initialize the Count Vectorizer
countvec_ge = CountVectorizer(max_df=.95,
                           min_df=.02,
                           stop_words = 'english')
```

```python
#create corpus of "documents"
documents_ge = msnbc_df_ge['title'].values

#fit model
cv_ge = countvec_ge.fit_transform(documents_ge)
cv_feature_names_ge = countvec_ge.get_feature_names_out()
```

```python
#create the LDA topic model
lda_ge = LatentDirichletAllocation(n_components=20, 
                                max_iter=100, 
                                batch_size=99,
                                learning_method='online').fit(cv_ge)
```

Below I used a function that I had previously used in a lab in a course in undergrad (Legal Studies 123) when using LDA to do topic modeling. The first argument is the LDA model above. The second argument is the list of words that label each column in the document matrix created in the line of code where I used get_feature_names_out to get the feature names from the document-term matrix I used to fit the topic model. The third argument is defined in the cell below, or the number of words I want returned for each topic.

The for loop goes over every topic in the model. "model.components_" is an array in which each row corresponds to a topic and each column corresponds to the word's importance in the topic. I then use argsort with slicing to get the indices of words in descending order and n_top_words to get the indices with the top 10 words. 

Then I print the topic number, and the final print line converts the indices back to the actual word in the topic. The output of applying the function then gives me 20 topics (as defined in the n_components argument of the LDA model) with the top ten words for each topic, in descending order.

I used this function because I wanted the output to be readable and easily interpretable as opposed to just displaying it in a matrix.

```python
num_top_words = 10
 
def topic_words(model, feature_names, n_top_words):
    for num_topic, topic in enumerate(model.components_):
        words = np.argsort(topic)[::-1][:n_top_words]
        print('Topic' + str(num_topic) + ':')
        print(' '.join([feature_names[i] for i in words]))

topics = topic_words(lda_ge, cv_feature_names_ge, num_top_words)
print(topics)
```

Unfortunately, the LDA did not yield a list of very distinct topics. I assume this is because the corpus of documents (headlines) was small. Perhaps if I was looking at a broader time frame and had a larger set of headlines, there would be a wider variety of articles yielding more distinct topics. ## TF-IDF

Below I use a TF-IDF Vectorizer to get the top phrases across headlines. I set the ngram range to (1,3) to consider unigrams, bigrams, and trigrams, or sequences of words that occur together across the documents, such as "covid 19" as opposed to just "covid", or "donald trump" as opposed to just "trump". I had to change the max_df a couple of times to get words that had a document frequency higher than 0.99. I originally had 0.95, but the output had some filler/non-corpus-specific stop words, so I increased it. I put the min_df as 3, which I also had to change a few times to get a list of top phrases that seemed to make some sense. 

```python
#make text list
text_list_ge = msnbc_df_ge['title']

#convert text data to tf-idf
tf_ge = TfidfVectorizer(analyzer='word', 
                     ngram_range=(1,3), #set this to 1,3 to consider unigrams, bigrams, and trigrams
                     max_df = 0.99,
                     min_df=3,
                     stop_words = ('english'))
#fit the model
tfidf_matrix_ge = tf_ge.fit_transform(text_list_ge)
#extract feature names
feature_names_ge = tf_ge.get_feature_names_out()
#check how many feature names I got
print(len(feature_names_ge))
                     
```

Below I used .todense() to make the TF-IDF matrix less sparse, since I was running into error codes with a sparse matrix. Then I summed the TF-IDF scores in order to determine the importance of the terms across all of the headlines published on MSNBC for the month of October 2020. Then I printed the top 20 phrases and their TF-IDF score for the dataset. 

Again, if the data had labels or if I had time to build/optimize a model that could tag the articles, I would use the TF-IDF scores as features for the dataset to build a predictive model that could potentially be used for prediction for another election year.

```python
dense_ge = tfidf_matrix_ge.todense()

#sum tf-idf scores for all the headlines
sum_tf_ge = np.array(tfidf_matrix_ge.sum(axis=0)).flatten()
phrase_scores_ge = [(feature_names_ge[i], sum_tf_ge[i]) for i in range(len(feature_names_ge))]

sorted_phrase_scores_ge = sorted(phrase_scores_ge, key=lambda t: t[1],reverse=True)

print(f"Top phrases across headlines:")
for phrase, score in sorted_phrase_scores_ge[:20]:
    print('{0: <20} {1}'.format(phrase, score))
```

The term "trump" had the highest TF-IDF score, which indicates that the term was important across the entire set of MSNBC headlines for October 2020. This makes because he was the opposing candidate in the election, but also because a huge part of the Democratic campaign in that election year was running on a "never Trump" platform. It would seem logical the MSNBC would run articles that were critical of Trump, his policies, and his campaign for presidency, and that lends itself to making the term more important in the corpus. "msnbc" as a term also had a high score, but contextually I can interpret this as being because the term appears when referring to their own exclusive report in the headline or something similar. "women", "covid", and "black" also appeared on the higher end of TF-IDF scores, and when compared to other words on the list, can be taken as potential political topics or coming from articles that may be discussing a particular political issue. "women" likely appears with a higher score because there were candidates identifying as women running, and women's rights were also a concern given the conservative majority on the Supreme Court and Trump's behavior towards women. "black" and "covid" likely had higher scores because of the prevalence of pandemic related issues and concerns about racial inequality, and potential higher media coverage of these issues leading up to the election.

Below I apply the same functions as above for MSNBC articles published in October 2022, the month preceding the 2022 Midterm Elections. I created a dataframe to store the article titles and urls.

```python
#using mt to show this is info from month before 2022 midterm election
msnbc_url_mt = 'https://www.msnbc.com/archive/articles/2022/october'

#apply the function we made earlier
msnbc_soup_mt = get_soup(msnbc_url_mt)

#select class a with href attribute (where link and title contained in html)
msnbc_a_mt = msnbc_soup_mt.select('a[href]')

#initialize dictionary to store output
msnbc_articles_mt = {}

#for loop to append the links and titles to the dict
for link in msnbc_a_mt:
        href = link.get('href')
        if "https://" not in href: #getting rid of as many pages as possible that aren't articles
            continue
        title = link.get_text(strip=True)
        msnbc_articles_mt[href]=title

#convert dictionary to dataframe to use for topic modeling
msnbc_df_mt = pd.DataFrame.from_dict(msnbc_articles_mt, orient='index').reset_index()
msnbc_df_mt = msnbc_df_mt.rename(columns={"index": "url", 0:"title"})
msnbc_df_mt
```

## NLP

Below I use simple NLP to get the counts of the most common words across all the headlines published on MSNBC in October 2022. 

```python
nlp_mt = spacy.load("en_core_web_sm")
nlp_mt.add_pipe("spacytextblob")

all_headlines_mt = ''.join(msnbc_df_mt['title'])

doc_mt = nlp_mt(all_headlines_mt)

#tokenize, lower case, remove stopwords/punctuation
tokens_mt = [token.text.lower() for token in doc_mt if not token.is_stop and token.is_alpha]

word_counts_mt = Counter(tokens_mt)

word_counts_mt.most_common(20)
```

Again, "trump" is the most commonly used term across the headlines in the dataset. "walker" is likely related to Herschel Walker, who ran in the contentious 2022 Senate election in Georgia. That specific race received a lot of coverage at the time, since Walker was a very controversial and conservative Trump-backed candidate, and his opposition, Raphael Warnock, was an anti-Trump Democrat. Other specific terms that appear in this list are "supreme" and "court", likely because this was the first election after the Supreme Court published their opinion overruling Roe v. Wade, endangering rights to abortion; "abortion" and "women" also appear in the count of most common words across the headlines. "black" has a count of 8, which indicates that possibly either race-related issues or Black electoral candidates were highlighted on MSNBC in October 2022. 

```python
#get the polarity for each one of these articles, add as new column
msnbc_df_mt = msnbc_df_mt.copy()

msnbc_df_mt.loc[:,'polarity'] = msnbc_df_mt['title'].apply(lambda text: nlp_mt(text)._.polarity)

#get the subjectivity for each one of these articles, add as new column
msnbc_df_mt = msnbc_df_mt.copy()

msnbc_df_mt.loc[:, 'subjectivity'] = msnbc_df_mt['title'].apply(lambda text: nlp_mt(text)._.subjectivity)

msnbc_df_mt
```

## TF-IDF Vectorizer

Below I use a TF-IDF vectorizer, similar to the one I used for the MSNBC data for the month preceding the General Election in 2020.

```python
#make text list
text_list_mt = msnbc_df_mt['title']

#convert text data to tf-idf
tf_mt = TfidfVectorizer(analyzer='word', 
                     ngram_range=(1,3), #set this to 1,3 to consider unigrams, bigrams, and trigrams
                     max_df = 0.99,
                     min_df=3,
                     stop_words = ('english'))

tfidf_matrix_mt = tf_mt.fit_transform(text_list_mt)
feature_names_mt = tf_mt.get_feature_names_out()
print(len(feature_names_mt))
                     
```

```python
dense_mt = tfidf_matrix_mt.todense()

#sum tf-idf scores for all the headlines
sum_tf_mt = np.array(tfidf_matrix_mt.sum(axis=0)).flatten()
phrase_scores_mt = [(feature_names_mt[i], sum_tf_mt[i]) for i in range(len(feature_names_mt))]

sorted_phrase_scores_mt = sorted(phrase_scores_mt, key=lambda t: t[1],reverse=True)

print(f"Top phrases across headlines:")
for phrase, score in sorted_phrase_scores_mt[:20]:
    print('{0: <20} {1}'.format(phrase, score))
```

The terms likely related to some political issue with the highest TF-IDF scores are "trump", "jan" (for January 6th, likely), "walker", "biden", and "herschel walker". 

For readers of MSNBC answering exit polls, this could indicate that they might answer that beating the Republican party for the way they handled Trump's general election loss (January 6th) and their campaigning for Herschel Walker were considerations at top of mind.

## MeanShift Clustering Using General Election TF-IDF Vectorizer

```python
#import packages
from sklearn.cluster import MeanShift, estimate_bandwidth

#bandwidth isn't given so using this function from sklearn
#using .toarray because error message said to make it dense
bandwidth = estimate_bandwidth(tfidf_matrix_ge.toarray(), quantile=0.2, n_samples=500)

#MeanShift Clustering
ms = MeanShift(bandwidth=bandwidth, bin_seeding=True)

#fit model
ms.fit(tfidf_matrix_ge.toarray())
labels = ms.labels_
cluster_centers = ms.cluster_centers_

labels_unique = np.unique(labels)
n_clusters_ = len(labels_unique)

print("num est clusters : %d" % n_clusters_)
```

```python
import matplotlib.pyplot as plt

plt.figure(1)
plt.clf()

colors = ["#dede00", "#377eb8", "#f781bf"]
markers = ["x", "o", "^"]

X = tfidf_matrix_ge.toarray()

for k, col in zip(range(n_clusters_), colors):
    my_members = labels == k
    cluster_center = cluster_centers[k]
    plt.plot(X[my_members, 0], X[my_members, 1], markers[k], color=col)
    plt.plot(
        cluster_center[0],
        cluster_center[1],
        markers[k],
        markerfacecolor=col,
        markeredgecolor="k",
        markersize=14,
    )
plt.title("Estimated number of clusters: %d" % n_clusters_)
plt.show()
```

The MeanShift clustering indicates that there are 48 clusters, which could be due to the data having a lot of noise or having many outliers, which leads to there being more individual clusters than a few significant clusters. Even with estimating the bandwidth, there were a lot of clusters. It is hard to interpret the significance of this MeanShift clustering, and it probably doesn't add anything helpful to my analysis. 

## VBGMM Clustering Using General Election TF-IDF Vectorizer

```python
#import packages
from sklearn.mixture import BayesianGaussianMixture
from sklearn.decomposition import PCA

#reduce dimensionality
data = tfidf_matrix_ge.toarray()

#reduce dimensionality with scikitlearn PCA
if data.shape[1] > 2:
    pca = PCA(n_components=2)
    reduced_data = pca.fit_transform(data)
else:
    reduced_data = data

#fit the model
vbgmm = BayesianGaussianMixture(n_components=6, random_state=2) #set max number of clusters to 6 (chosen at random)
vbgmm.fit(data)
labels_v = vbgmm.predict(data)

#plot with matplotlib
plt.figure()
labels_uq = np.unique(labels_v)
colors = plt.cm.viridis(np.linspace(0, 1, len(labels_uq)))

#each term is colored depending on which cluster it is in (Chat GPT transcript included for this cell)
for k, col in zip(labels_uq, colors):
    class_members = (labels_v == k)
    plt.scatter(reduced_data[class_members, 0], reduced_data[class_members, 1], color=col, label=f'Cluster {k}')
    #overlaying the terms in the cluster on the plot
    plt.text(reduced_data[class_members, 0].mean(), 
             reduced_data[class_members, 1].mean(), 
             f'{sorted_phrase_scores_ge[k]}',
             fontsize=12, color='black', weight='bold')

#print each term and its assigned cluster
for term, label in zip(sorted_phrase_scores_ge, labels_v):
    print(f'Term: {term}, Cluster: {label}')

plt.title('VBGMM Clustering Results')
plt.xlabel('Component 1')
plt.ylabel('Component 2')
plt.legend()
plt.show()
```

```python
#store the list of terms and their cluster as tuples because that list is long and hard to process
tuples = [(term, label) for term, label in zip(sorted_phrase_scores_ge, labels_v)]
cluster_df = pd.DataFrame(tuples, columns=['term', 'cluster'])
#group to get a sense of how many terms fall into a cluster
cluster_df.groupby(by='cluster').count()
```

The VBGMM clustering shows six clusters that have been formed, which tracks based on setting n_components = 6. The axes show Component 1 and Component 2, and that is because I used PCA to reduce the dimensionality of the data. There is a large cluster for cluster 1 but some of the points are dispersed over the y and x axes of the plot. There is also overlap between clusters, which is indicative of terms that could belong to multiple clusters. There is a discrepancy between the number of clusters visible on the plot and the number of clusters that show up in the groupby dataframe above. The length of the feature_names_ge from the TF-IDF vectorization is 55, so there were 55 unique terms from the TF-IDF vectorization. 49 of those terms fall into cluster 1, 1 falls into cluster 0, and 5 fall into cluster 4. I presume this is because there is overlap between terms in clusters, and 49 of the terms belong to cluster 1 but could also be represented as belonging to the other clusters. To me, this also demonstrates the inefficacy of using machine learning models on this particular set of data, supervised or not, as this doesn't tell us much about the data-- cluster 1 is a pretty wide net to cast for these terms.

As stated earlier in this notebook, the dataset isn't large enough to develop distinct topics from topic models, or to use for robust prediction models. This is partially due to my selection method and the parameters I chose to work under, but also possibly due to the narrowing down of topics MSNBC chose to focus on before the election.
